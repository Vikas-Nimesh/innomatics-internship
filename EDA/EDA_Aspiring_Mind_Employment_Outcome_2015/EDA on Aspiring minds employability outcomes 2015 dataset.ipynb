## `About Dataset`

The dataset was released by Aspiring Minds from the Aspiring Mind Employment Outcome 2015 (AMEO). <br>
The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary, Job Titles, and Job Locations) along with the standardized scores from three different areas – cognitive skills, technical skills and personality skills. The dataset also contains demographic features. The dataset  contains  around  40 independent variables and 4000 data points. The independent variables are both continuous and categorical in nature. The dataset contains a unique identifier for each candidate. Below mentioned table contains the details for the original dataset.

DOL (Date of Joining) have incorrect datatype and present some string values.

```python
df['DOL'].replace('present',pd.Timestamp(pd.to_datetime("today").date()),inplace=True)
```

# `Exploratory Data Analysis(EDA)`

How to choose the right visualization method?<br>
When visualizing individual variables, it is important to first understand what type of variable you are dealing with. This will help us find the right visualization method for that variable.

**`Spliting of numerical and categorical columns`**

```python
num_col = df.select_dtypes(include= ['float','int']).columns
cat_col = df.select_dtypes(include= ['object']).columns
```

```python
num_col,cat_col
```

## `Outlier handling`

```python
q1 = df[num_col].quantile(0.25)
q3 = df[num_col].quantile(0.75)
IQR = q3 - q1
```

```python
IQR
```

```python
min_value = q1 - (1.5 * IQR)
max_value = q3 + (1.5 * IQR)
```

```python
print('Outliers below:\n',df[num_col].lt(q1 - 1.5*IQR).sum())
print()
print('Outliers above:\n',df[num_col].gt(q3 + 1.5*IQR).sum())
```

**`Boxplots`**

```python
len(num_col)
```

```python
plt.figure(figsize=(20,60))

for i, col in enumerate(num_col):
    plt.subplot(15,2,i+1)
    sns.boxplot(df[col])
    
    plt.xlabel(col,fontsize = 15)
    plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
```

```python
plt.figure(figsize=(20,60))

for i, col in enumerate(num_col):
    plt.subplot(15,3,i+1)
    sns.distplot(df[col],hist = False)
    plt.ylabel('Density',fontsize = 20)
    plt.xlabel(col,fontsize = 20)
    plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
```

```python
plt.figure(figsize=(20,60))
for i, col in enumerate(num_col):
    plt.subplot(15,3,i+1)
    plt.hist(df[col])
    plt.xlabel(col,fontsize = 20)
    plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
```

### `Bivariate  Analysis`:<br>
When we talk about bivariate analysis, it means analyzing 2 variables. Since we know there are numerical and categorical variables, there is a way of analyzing these variables as shown below:
<br><br>
i) Numerical & Numerical<br>
   a. Scatterplot<br>
   b. Line plot<br>
   c. Heatmap for correlation<br><br>
ii) Numerical & Categorical<br>
   a. Bar chart<br>
   b. Categorical box plot<br><br>
iii) Categorical & Categorical<br>
   a. Bar chart<br>
   b. Grouped bar chart<br>

```python
plt.figure(figsize = (25,50))

for i,col in enumerate(num_col[1:]):
    plt.subplot(10,3,i+1)
    plt.ticklabel_format(style='plain')
    sns.scatterplot(x = col, y = 'Salary', data = df)
    plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
plt.show()
```

```python
# Gender vs Salary
plt.figure(figsize = (15, 4))
plt.subplot(1, 2, 1)
sns.barplot(x = df['Gender'], y = df['Salary'],ci=0)
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Gender'], y = df['Salary'])
plt.show()
```

```python
# CollegeTier vs Salary
plt.figure(figsize = (15, 4))
plt.subplot(1, 2, 1)
sns.barplot(x = df['CollegeTier'], y = df['Salary'],ci=0)
plt.subplot(1, 2, 2)
sns.boxplot(x=df['CollegeTier'], y = df['Salary'])
plt.show()
```

```python
#  we can also get teh following conclusion from multi variate plot
plt.figure(figsize = (15, 4))
plt.subplot(1, 2, 1)
sns.barplot(x = df['CollegeTier'], y = df['Salary'], hue = df['Gender'],ci=0)
plt.subplot(1, 2, 2)
sns.boxplot(x = df['CollegeTier'], y = df['Salary'], hue = df['Gender'])
plt.show()
```

```python
plt.figure(figsize = (17,5))
plt.subplot(1,2,1)
sns.barplot(data = df , x = 'GraduationYear',y = 'Salary',ci=0)
plt.subplot(1,2,2)
sns.barplot(data = df , x = '12graduation',y = 'Salary',ci=0)
plt.xlabel('Passing Year of 12th')
```
